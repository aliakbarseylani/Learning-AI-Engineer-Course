## Draw a line plot (graph).

```python
season_x = ['Spring', 'Summer', 'Autumn', 'Winter']
temperature_y = [20, 37, 18, 4]

plt.xlabel('Season')
plt.ylabel('Temperature')
plt.title('Change Weather Between Season')
plt.plot(season_x,temperature_y, marker = 'o')
plt.show()
```

## 1. Changes Weather Between Month in year 2024 Mashhad, Iran

```python
months_x = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
temperature_y = [9, 10, 16, 22, 28, 34, 35, 34, 30, 23, 16, 11]

plt.xlabel('Months')
plt.ylabel('Temperature')
plt.title('Changes Weather Between Month in year 2024 Mashhad, Iran')
plt.plot(months_x, temperature_y, marker = 'o', alpha=0.5, color = 'blue', linewidth=2)
plt.xticks(rotation = 45 )
plt.show()
```

## 2. draw a bar chart that show number of student in three different course at 5 years.

```python
import numpy as np

years = (1400, 1401, 1402, 1403, 1404)
x_indexes = np.arange(len(years))
                      
chemistry_student = (30, 31, 32, 29, 35)
math_students = (25, 27, 26, 28, 29) 
physics_students = (40, 41, 42, 44, 45)

width = 0.25

plt.bar(x_indexes - width, chemistry_student, width = 0.25, label='chemistry_student')
plt.bar(x_indexes, math_students, width = 0.25, label= 'math_students')
plt.bar(x_indexes + width, physics_students, width = 0.25, label='physics_students')
plt.xticks(ticks=x_indexes, labels=years)
plt.legend()
plt.tight_layout()
plt.show()

```

## 3. show the share of sales of 5 products at a shop in a month.

```python
products = ['Cloth', 'Farnicher', 'Book', 'Mobile', 'Laptop']
sales = [30, 80, 15, 40, 90]

explode = [0, 0, 0.1, 0 ,0]

plt.figure(figsize = (8,8))
plt.pie(sales, labels=products, autopct ='%1.f%%', explode=explode)
plt.show()
```
