## 8. Calculating the average total ADR (With two decimal places)

## 12. countries had most children and babies

```python
country_children_babies = hb.groupby('country')[['children', 'babies']].sum()

country_children_babies['total_children_babies'] = country_children_babies['children'] + country_children_babies['babies']

top_countries = country_children_babies.sort_values('total_children_babies', ascending=False).head(10)

print("کشورهایی با بیشترین تعداد کودکان و نوزادان در رزرو هتل:")
print(top_countries)
```
